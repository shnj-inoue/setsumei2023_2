以下のプログラムはBINGOができるプログラムです。

```python
import numpy as np
import random

def rand_nodup(a, b, k):
    r = set()
    while len(r) < k:
        r.add(random.randint(a, b))
    r=list(r)
    random.shuffle(r)
    return r

def display_card(card):
    for row in card:
        print("|"+ "|".join(f"{num:2}" if num != 0 else "  " for num in row)+"|")

number=[]
t=0

card = np.zeros((5,5))
card[0] = np.array(rand_nodup(1, 15, 5))
card[1] = np.array(rand_nodup(16, 30, 5))
card[2] = np.array(rand_nodup(31, 45, 5))
card[3] = np.array(rand_nodup(46, 60, 5))
card[4] = np.array(rand_nodup(61, 75, 5))
card[2,2]=0
card = card.astype(int)
card = np.array(card).T
print("ビンゴカード↓")
display_card(card)

while True:

  sum1=0
  sum2=0
  a=0
  n=0
  t+=1

  while True:
    n = np.random.randint(1,76)
    if n not in number:
      number.append(n)
      print(f"{n}が出ました")
      break

  if n in card:
    print(f"{n}が開きます")
    card=np.where(card==n,0,card)
    display_card(card)
  else:
    print(f"{n}はありませんでした")
    display_card(card)

  for i in range(5):
    if np.sum(card[i,:])==0:
      a=1
      break
    elif  np.sum(card[:,i])==0:
      a=1
      break
    sum1+=card[i,i]
    if i==4:
      if sum1==0:
        a=1
        break
    sum2+=card[i,4-i]
    if i==4:
      if sum2==0:
        a=1
        break
  if a==1:
    print(f"BINGO!!\n{t}回目でビンゴしました")
    break
```
